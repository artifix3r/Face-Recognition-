```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.datasets import fetch_lfw_people
```

```python
#face recognition is the supervised classification taks of identifying a person who is an image her or his face. In this project we used lfw people data set
```

```python
#loading data 
data=fetch_lfw_people(min_faces_per_person=70, resize=0.4)
```

```python

X,y=data.data,data.target
```

```python
#we then randomly split the images into training and test sets 
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=11)
```

```python
#and fit the PCA on the training set
pca=PCA(n_components=150,whiten=True)
```

```python
#We reduce all the instances to 150 dimensions and train logistic regression classifier. The dataset contains 40 classes. 
#Scikit-learn automatically creates binary and classifiers using the one-versus-all strategy behind the scenes  
X_train_PCA=pca.fit_transform(X_train)
X_test_PCA=pca.transform(X_test)
print(X_train.shape)
print(X_train_PCA.shape)
```

```python
classifier=LogisticRegression()
accuracies=cross_val_score(classifier,X_train_PCA,y_train)
```

```python
print('Cross validation accuracy: %s' % np.mean(accuracies))
```

```python
classifier.fit(X_train_PCA,y_train)
predictions=classifier.predict(X_test_PCA)
print(classification_report(y_test,predictions))
```
